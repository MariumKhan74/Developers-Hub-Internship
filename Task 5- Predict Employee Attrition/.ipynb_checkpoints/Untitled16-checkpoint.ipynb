```python
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(data, test_size=0.2, stratify=data['Attrition'], random_state=42)
```

```python
train_data['Attrition_Yes'] = (train_data['Attrition'] == 'Yes').astype(int)
test_data['Attrition_Yes'] = (test_data['Attrition'] == 'Yes').astype(int)
```

```python
train_encoded = pd.get_dummies(train_data, drop_first=True)
test_encoded = pd.get_dummies(test_data, drop_first=True)
```

```python
test_encoded = test_encoded.reindex(columns=train_encoded.columns, fill_value=0)
```

```python
X_train = train_encoded.drop('Attrition_Yes', axis=1)
y_train = train_encoded['Attrition_Yes']

X_test = test_encoded.drop('Attrition_Yes', axis=1)
y_test = test_encoded['Attrition_Yes']
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```
